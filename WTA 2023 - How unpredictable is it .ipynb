```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wta = pd.read_csv('wta_matches_2023.csv')

wta.head()
```

```python
wta.columns
```

```python
plt.figure(figsize = (25,10))
sns.countplot(x='winner_ioc', data=wta)
plt.savefig('wta2023_1.png')
```

```python
wta['winner_ioc'].unique()
```

```python
wta['winner_ioc'].value_counts()
```

```python
wta['upset'] = np.where(wta['loser_rank'] > wta['winner_rank'], 'Yes', 'No')
```

```python
wta.head()
```

```python
# Count occurrences of 'Yes' and 'No'
count_result = wta['upset'].value_counts()

# Plot a pie chart
plt.pie(count_result, labels=count_result.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Did the higher ranked player win??')
plt.show()
```

```python
grandslam = wta[wta['tourney_level'] == 'G']
# Count occurrences of 'Yes' and 'No'
count_result = grandslam['upset'].value_counts()

# Plot a pie chart
plt.pie(count_result, labels=count_result.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Did the higher ranked player win?? (at GrandSlams)( WTA)')
plt.show()
```
